```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import missingno as msno
import matplotlib.pyplot as plt
# pd.set_option('display.max_rows', None)
```

```python
df = pd.read_excel("qry_ept.xlsx")
df.head()
```

```python
msno.matrix(df)
plt.title('Missing Values Matrix')
plt.show()
```

```python
# Handle NA
df = df.dropna(subset=["TotalSkor"])
msno.matrix(df)
plt.title('Missing Values Matrix')
plt.show()
```

```python
# Filter tabel1 hanya yang Mahasiswa Polnep Semester Akhir dan Awal
tabel1 = df[(df["Kegiatan"] == "Semester Akhir") | (df["Kegiatan"] == "Semester Awal")]


tabel1 = tabel1[tabel1["Status"] == "Mahasiswa"]
tabel1 = tabel1[tabel1["Unit"] == "Polnep"]
tabel1 = tabel1[["Tahun","Diploma", "Jurusan","Prodi_2", "Kegiatan", "TotalSkor"]]
# Visualize missing values using missingno
msno.matrix(tabel1)
plt.title('Missing Values Matrix')
plt.show()
```

```python
# Membuat tabel perbandingan nilai rata-rata semester awal dan akhir

tbl_1_semakhir = tabel1[tabel1["Kegiatan"] == "Semester Akhir"]
tbl_1_semakhir_summary = tbl_1_semakhir.groupby("Tahun")
tbl_1_semakhir_summary = tbl_1_semakhir_summary ["TotalSkor"].agg(["count", "mean"]).reset_index()

tbl_1_semawal = tabel1[tabel1["Kegiatan"] == "Semester Awal"]
tbl_1_semawal_summary  = tbl_1_semawal.groupby("Tahun")
tbl_1_semawal_summary  = tbl_1_semawal_summary ["TotalSkor"].agg(["count", "mean"]).reset_index()

tabel_mean_summary = pd.merge(tbl_1_semakhir_summary, tbl_1_semawal_summary, on="Tahun", how="outer")
tabel_mean_summary = tabel_mean_summary.rename(columns={
    "count_x":"jml_sem_akhir",
    "mean_x": "mean_sem_akhir",
    "count_y":"jml_sem_awal",
    "mean_y": "mean_sem_awal",
})
tabel_mean_summary
```

```python
# Pembuatan tabel persentase secara global

## semester akhir
tabel_persentase_sem_akhir = tbl_1_semakhir.groupby("Tahun").agg(
    total_mhs = ("TotalSkor", "count"),
    jml_mhs_diatas_400 = ("TotalSkor", lambda x: (x >= 400).sum()),
    jml_mhs_dibawah_400 = ("TotalSkor", lambda x: (x < 400).sum()),
)
tabel_persentase_sem_akhir["persentase"] = tabel_persentase_sem_akhir["jml_mhs_diatas_400"] / tabel_persentase_sem_akhir["total_mhs"]

## semester awal
tabel_persentase_sem_awal = tbl_1_semawal.groupby("Tahun").agg(
    total_mhs = ("TotalSkor", "count"),
    jml_mhs_diatas_400 = ("TotalSkor", lambda x: (x >= 400).sum()),
    jml_mhs_dibawah_400 = ("TotalSkor", lambda x: (x < 400).sum()),
)
tabel_persentase_sem_awal["persentase"] = tabel_persentase_sem_awal["jml_mhs_diatas_400"] / tabel_persentase_sem_awal["total_mhs"]

```

```python
# tabel mean semester akhir per diploma
tbl_1_semakhir_summary_per_diploma = tbl_1_semakhir.groupby(["Tahun", "Diploma"]).agg(
    total_mhs = ("TotalSkor", "count"),
    rata_rata = ("TotalSkor", "mean"),
    ge_400 = ("TotalSkor", lambda x: (x >= 400).sum()),
    lt_400 = ("TotalSkor", lambda x: (x < 400).sum()),
)
tbl_1_semakhir_summary_per_diploma["persentase"] = tbl_1_semakhir_summary_per_diploma["ge_400"] / tbl_1_semakhir_summary_per_diploma["total_mhs"]


# tabel mean semester akhir per jurusan
tbl_1_semakhir_summary_per_jurusan = tbl_1_semakhir.groupby(["Tahun", "Jurusan"]).agg(
    total_mhs = ("TotalSkor", "count"),
    rata_rata = ("TotalSkor", "mean"),
    ge_400 = ("TotalSkor", lambda x: (x >= 400).sum()),
    lt_400 = ("TotalSkor", lambda x: (x < 400).sum()),
)
tbl_1_semakhir_summary_per_jurusan["persentase"] = tbl_1_semakhir_summary_per_jurusan["ge_400"] / tbl_1_semakhir_summary_per_jurusan["total_mhs"]

# tabel mean semester akhir per prodi
tbl_1_semakhir_summary_per_prodi = tbl_1_semakhir.groupby(["Tahun", "Prodi_2"]).agg(
    total_mhs = ("TotalSkor", "count"),
    rata_rata = ("TotalSkor", "mean"),
    ge_400 = ("TotalSkor", lambda x: (x >= 400).sum()),
    lt_400 = ("TotalSkor", lambda x: (x < 400).sum()),
)
tbl_1_semakhir_summary_per_prodi["persentase"] = tbl_1_semakhir_summary_per_prodi["ge_400"] / tbl_1_semakhir_summary_per_prodi["total_mhs"]
```

```python
# tabel mean semester awal per diploma
tbl_1_semawal_summary_per_diploma = tbl_1_semawal.groupby(["Tahun", "Diploma"]).agg(
    total_mhs = ("TotalSkor", "count"),
    rata_rata = ("TotalSkor", "mean"),
    ge_400 = ("TotalSkor", lambda x: (x >= 400).sum()),
    lt_400 = ("TotalSkor", lambda x: (x < 400).sum()),
)
tbl_1_semawal_summary_per_diploma["persentase"] = tbl_1_semawal_summary_per_diploma["ge_400"] / tbl_1_semawal_summary_per_diploma["total_mhs"]


# tabel mean semester awal per jurusan
tbl_1_semawal_summary_per_jurusan = tbl_1_semawal.groupby(["Tahun", "Jurusan"]).agg(
    total_mhs = ("TotalSkor", "count"),
    rata_rata = ("TotalSkor", "mean"),
    ge_400 = ("TotalSkor", lambda x: (x >= 400).sum()),
    lt_400 = ("TotalSkor", lambda x: (x < 400).sum()),
)
tbl_1_semawal_summary_per_jurusan["persentase"] = tbl_1_semawal_summary_per_jurusan["ge_400"] / tbl_1_semawal_summary_per_jurusan["total_mhs"]


# tabel mean semester awal per prodi
tbl_1_semawal_summary_per_prodi = tbl_1_semawal.groupby(["Tahun", "Prodi_2"]).agg(
    total_mhs = ("TotalSkor", "count"),
    rata_rata = ("TotalSkor", "mean"),
    ge_400 = ("TotalSkor", lambda x: (x >= 400).sum()),
    lt_400 = ("TotalSkor", lambda x: (x < 400).sum()),
)
tbl_1_semawal_summary_per_prodi["persentase"] = tbl_1_semawal_summary_per_prodi["ge_400"] / tbl_1_semawal_summary_per_prodi["total_mhs"]
tbl_1_semawal_summary_per_prodi
```

```python
# daftar tabel:
# df
ept = {"global": tabel1, # tabel hanya mahasiswa polnep semester akhir dan awal
"semakhir": tbl_1_semakhir,
"semawal": tbl_1_semawal,
"mean_summary": tabel_mean_summary,
"persentase_sem_akhir": tabel_persentase_sem_akhir,
"persentase_sem_awal": tabel_persentase_sem_awal,
"semakhir_sum_per_diploma": tbl_1_semakhir_summary_per_diploma,
"semakhir_sum_per_jurusan": tbl_1_semakhir_summary_per_jurusan,
"semakhir_sum_per_prodi": tbl_1_semakhir_summary_per_prodi,
"semawal_sum_per_diploma": tbl_1_semawal_summary_per_diploma,
"semawal_sum_per_jurusan": tbl_1_semawal_summary_per_jurusan}

with pd.ExcelWriter("ept.xlsx") as writer:
    for sheet_name, df in ept.items():        
        df.to_excel(writer, sheet_name=sheet_name, index=False)
```
